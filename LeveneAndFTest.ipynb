# **Levene's Test and F-Test**

### **Null Hypothesis**: There is no significant difference between the variances of Groups A and B

### **Alternative Hypothesis**: There is a significant difference between the variances of Groups A and B

```python
import pandas as pd
import numpy as np
from scipy.stats import levene, f
```

```python
df = pd.read_csv("Customer Satisfaction Scores and Behavior Data.csv")
df.head()
```

```python
data = df[["Group", "Satisfaction_Score"]]
data.head()
```

```python
data["Group"].unique()
```

```python
a_group = np.array(data[data["Group"] == 'A']['Satisfaction_Score'])
b_group = np.array(data[data["Group"] == 'B']['Satisfaction_Score'])
```

```python
stat, p = levene(a_group, b_group)
print("Levene Statistic: ", stat)
print("p value: ", p)
```

```python
alpha = 0.05

if p < alpha:
    print("Rejected Null Hypothesis")
    print("There is a significant difference between the variances of Groups A and B")
else:
    print("Failed to Reject Null Hypothesis")
    print("There is no significant difference between the variances of Groups A and B")
```

```python
a_var = np.var(a_group, ddof = 1)
b_var = np.var(b_group, ddof = 1)

print("Variance of Group A: ", a_var)
print("Variance of Group B: ", b_var)
```

```python
if a_var > b_var:
    F = a_var / b_var
    dfn = len(a_group) - 1
    ddn = len(b_group) - 1
else:
    F = b_var / a_var
    dfn = len(a_group) - 1
    ddn = len(b_group) - 1

print("F-statistic: ", F)
```

```python
p_value = 2 * min(
    f.cdf(F, dfn, ddn),
    1 - f.cdf(F, dfn, ddn)
)

print("p value: ", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Rejected Null Hypothesis")
    print("There is a significant difference between the variances of Groups A and B")
else:
    print("Failed to Reject Null Hypothesis")
    print("There is no significant difference between the variances of Groups A and B")
```
