# **One Way ANOVA**

### **Null Hypothesis**: There is no significant difference between satisfaction scores of customers of different loyalty levels.

### **Alternative Hypothesis**: There is a significant difference between satisfaction scores of customers of different loyalty levels.

```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
```

```python
df = pd.read_csv("Customer Satisfaction Scores and Behavior Data.csv")
df.head()
```

```python
data = df[["Loyalty_Level", "Satisfaction_Score"]]
data.head()
```

```python
data["Loyalty_Level"].unique()
```

```python
low_sat = np.array(data[data["Loyalty_Level"] == "Low"]["Satisfaction_Score"])
medium_sat = np.array(data[data["Loyalty_Level"] == "Medium"]["Satisfaction_Score"])
high_sat = np.array(data[data["Loyalty_Level"] == "High"]["Satisfaction_Score"])
```

```python
f_stat, p_value = f_oneway(
    low_sat,
    medium_sat,
    high_sat
)

print("f statistic: ", f_stat)
print("p value: ", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Null Hypothesis Rejected")
    print("There is a significant difference between the satisfaction scores of customers of different loyalty levels")
else:
    print("Failed to Reject Null Hypothesis")
    print("There is no signigicant difference between the satisfaction scores of customers of different loyalty levels")
```
