```python
import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import ttest_ind
```

```python
df = pd.read_csv("Customer Satisfaction Scores and Behavior Data.csv")
df.head()
```

```python
group_a = np.array(df[df["Group"] == "A"]["Satisfaction_Score"])
group_b = np.array(df[df["Group"] == "B"]["Satisfaction_Score"])
```

```python
t_stat, p_value, dof = ttest_ind(
    group_a,
    group_b,
    usevar = "unequal"
)

print("T Statistic: ", t_stat)
print("P Value: ", p_value)
print("Degrees of Freedom: ", dof)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Rejected Null Hypothesis")
    print("There is a statistically significant difference between the average satisfaction scores of customers og Groups A and B")
else:
    print("Failed to Reject Null Hypothesis")
    print("There is no statistically significant difference between the average satisfaction scores of customers of Groups A and B")
```
