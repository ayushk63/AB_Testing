```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
```

```python
df = pd.read_csv("Customer Satisfaction Scores and Behavior Data.csv")
df.head()
```

```python
data = df[['Group', 'Satisfaction_Score']]
data.head()
```

```python
a_sat = np.array(data['Satisfaction_Score'][data['Group'] == 'A'])
b_sat = np.array(data['Satisfaction_Score'][data['Group'] == 'B'])
```

```python
t_stat, p_value = ttest_ind(
    a_sat,
    b_sat,
    equal_var = False
)

print("t statistic: ", t_stat)
print("p value: ", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Null Hypothesis Rejected")
    print("There is a significant difference between the average satisfaction scores of Groups A and B")
else:
    print("Failed to Reject Null Hypothesis")
    print("There is no significant difference between the average satisfaction scores of Groups A and B")
```
