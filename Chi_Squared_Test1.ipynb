```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv("Customer Satisfaction Scores and Behavior Data.csv")
df.head()
```

```python
data = df[['Gender', 'Purchase_History']]
data.head()
```

```python
df1 = pd.DataFrame()
df1['Gender'] = ['Male', 'Female']
df1['Purchase_History_Yes'] = [
    len(
        data[(data['Purchase_History'] == 'Yes') & (data['Gender'] == 'Male')]
    ),
    len(
        data[(data['Purchase_History'] == 'Yes') & (data['Gender'] == 'Female')]
    )
]
df1['Purchase_History_No'] = [
    len(
        data[(data['Purchase_History'] == 'No') & (data['Gender'] == 'Male')]
    ),
    len(
        data[(data['Purchase_History'] == 'No') & (data['Gender'] == 'Female')]
    )
]
df1.head()
```

```python
table = pd.crosstab(
    data['Gender'],
    data['Purchase_History']
)

print(table)
```

```python
chi2, p, dof, expected = chi2_contingency(table)

expected_df = pd.DataFrame(
    expected,
    index = table.index,
    columns = table.columns
)

print(expected_df)
```

```python
print("Degree of Freedom: ", dof)
print("p value: ", p)
print("chi2: ", chi2)
```

```python
alpha = 0.05

if (p < 0.05):
    print("Null Hypothesis Rejected")
    print("Gender and purchase history are dependent")
else:
    print("Failed to reject Null Hypothesis")
    print("Gender and purchase history are independent")
```
